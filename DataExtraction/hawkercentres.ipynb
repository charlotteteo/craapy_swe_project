```python
import pandas as pd
import requests
```

```python
hc=pd.read_csv('list-of-government-markets-hawker-centres.csv')
hc=hc[['name_of_centre','location_of_centre']]
hc
```

```python
list=[]
dict={}
for x in range(len(hc)):
    try:
        dict={}
            
        r = requests.get('https://developers.onemap.sg/commonapi/search?searchVal='+hc["name_of_centre"][x]+"&returnGeom=Y&getAddrDetails=Y&pageNum=1")

        result=r.json()["results"]
        result=pd.DataFrame(result)
        y={}
        if  (result["LATITUDE"][0])!=None :
            y["latitude"]=result["LATITUDE"][0]
        if  (result["LONGITUDE"][0])!=None :
            y["longitude"]=result["LONGITUDE"][0]
        dict["Name"]=hc["name_of_centre"][x]
        dict["Coordinates"]=y
        dict["Address"]="https://www.google.com/maps?saddr=My+Location&daddr="+result["LATITUDE"][0]+","+result["LONGITUDE"][0]
        print(dict,",")
    except:
        continue
    
    
```
